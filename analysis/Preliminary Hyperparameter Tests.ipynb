```python
from collections import defaultdict

layers = [1, 2, 3, 4]
neurons = [128, 256, 512, 1024, 2048]
timesteps= [5, 10, 20, 50, 100, 200]


index = -1
with open('layer_hyper_testing.txt', 'r') as f:
    lines = f.readlines()
    
    layer_result_dictionary = defaultdict(list)
    for line in lines:
        epoch_array = line.split('\t')
        
        raw_epoch = epoch_array[0]
        epoch = int(raw_epoch.split(' ')[1])

        loss = epoch_array[1]
        f1 = float(epoch_array[2].split(' ')[-1][:-1])
        
        if(epoch == 0):
            index += 1
        layer_result_dictionary[layers[index]].append(f1)

index = -1
with open('neuron_hyper_testing.txt', 'r') as f:
    lines = f.readlines()
    
    neuron_result_dictionary = defaultdict(list)
    for line in lines:
        epoch_array = line.split('\t')
        
        raw_epoch = epoch_array[0]
        epoch = int(raw_epoch.split(' ')[1])

        loss = epoch_array[1]
        f1 = float(epoch_array[2].split(' ')[-1][:-1])
        
        if(epoch == 0):
            index += 1
        neuron_result_dictionary[neurons[index]].append(f1)

index = -1
with open('timestep_hyper_testing.txt', 'r') as f:
    lines = f.readlines()
    
    timesteps_result_dictionary = defaultdict(list)
    for line in lines:
        epoch_array = line.split('\t')
        
        raw_epoch = epoch_array[0]
        epoch = int(raw_epoch.split(' ')[1])

        loss = epoch_array[1]
        f1 = float(epoch_array[2].split(' ')[-1][:-1])
        
        if(epoch == 0):
            index += 1
        timesteps_result_dictionary[timesteps[index]].append(f1)
```

```python
%matplotlib inline

fig, ax = plt.subplots(1, 3)

for key, value in layer_result_dictionary.items():
    ax[0].plot(range(200), value, label=str(key))

ax[0].set_xlabel('Number of Epochs')
ax[0].set_ylabel('F1 Score')
ax[0].legend(title='Layers')
ax[0].set_title('F1 Training Curves for Varying\n Number of Layers (512 neurons, 50 time-steps)')

for key, value in neuron_result_dictionary.items():
    ax[1].plot(range(200), value, label=str(key))

ax[1].set_xlabel('Number of Epochs')
ax[1].set_ylabel('F1 Score')
ax[1].legend(title='Neurons')
ax[1].set_title('F1 Training Curves for Varying\n Number of Neurons (1 layer, 50 time-steps)')

for key, value in timesteps_result_dictionary.items():
    ax[2].plot(range(200), value, label=str(key))

ax[2].set_xlabel('Number of Epochs')
ax[2].set_ylabel('F1 Score')
ax[2].legend(title='Time steps')
ax[2].set_title('F1 Training Curves for Varying\n Number of Time Steps (1 layer, 512 neurons)')

import matplotlib
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(30, 7)

plt.show()
```
